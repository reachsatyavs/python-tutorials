```python
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} says Woof!"

rex = Dog("Rex")
print(rex.speak())   # Rex says Woof!
```

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(name):                      # no self parameter!
        return name +  " Woof!"

rex = Dog("Rex")

Dog.speak("rex")
rex.speak("rex")
```
