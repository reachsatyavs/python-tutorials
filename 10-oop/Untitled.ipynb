```python
class Book:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def getBook(self):
         print(self.name, self.price)



book1 = Book("batman", 10)
book1.getBook()

book2 = Book("superman", 10)
book2.getBook()

# book1 and book2 are object that are created using template Book ( class - blueprint )
```
